# The Beatles - Revolver

_Revolver_ is the seventh studio album by the English rock band *The Beatles*. It was released on 5 August 1966. The members of The Beatles were: <b>John Lennon</b>, <b>Paul McCartney</b>, <b>George Harrison</b>, and <b>Ringo Starr</b> :)
